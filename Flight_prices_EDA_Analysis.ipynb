### FEATURES
The various features of the cleaned dataset are explained below:

1. **Airline:** The name of the airline company operating the flight. It is a categorical feature.
2. **Date_of_Journey:** The date on which the flight is scheduled to depart. It is a date/time feature.
3. **Source:** The city from which the flight departs. It is a categorical feature.
4. **Destination:** The city where the flight will arrive. It is a categorical feature.
5. **Route:** The path taken by the flight, including any connecting stops. It is a categorical feature.
6. **Dep_Time:** The departure time of the flight from the source city. It is a time feature.
7. **Arrival_Time:** The arrival time of the flight at the destination city. It is a time feature.
8. **Duration:** The total travel time between the source and destination cities. It is a continuous feature.
9. **Total_Stops:** The number of stops between the source and destination cities. It is an ordinal categorical feature.
10. **Additional_Info:** Miscellaneous information about the flight such as meal availability, baggage policy, or aircraft type. It is a categorical feature.
11. **Price:** The cost of the flight ticket. It is the target variable and a continuous feature.


```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
#splitting date,month & year into diff columns
df['date'] = df['Date_of_Journey'].str.split('/').str[0]
df['month'] = df['Date_of_Journey'].str.split('/').str[1]
df['year'] = df['Date_of_Journey'].str.split('/').str[2]
df.head()
```

```python
df.info()
```

```python
#converting str to int
df['date'] = df['date'].astype(int)
df['month'] = df['month'].astype(int)
df['year'] = df['year'].astype(int)
```

```python
df.info()
```

```python
#Dropping Date_of_Journey column
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
#splitting Arrival_Time column into hour & min
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head()
```

```python
#converting them into int
df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_min'] = df['Arrival_min'].astype(int)
```

```python
#dropping Arrival_Time column
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.info()
```

```python
df['Dep_Time'] = pd.to_datetime(df['Dep_Time'], format='%H:%M').dt.time
```

```python
df.head()
```

```python
#These are the unique values we have in Total_Stops column
df['Total_Stops'].unique()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
#most frequent value in 'Total_Stops' column
df['Total_Stops'].mode()
```

```python
#converting categorical values into numerical in Total_Stops column
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0, '1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1})
```

```python
#Converting duration into total mins
def to_minutes(x):
    h, m = 0, 0
    if 'h' in x:
        h = int(x.split('h')[0])
    if 'm' in x:
        m = int(x.split('m')[0].split()[-1])
    return h * 60 + m

df['Duration'] = df['Duration'].apply(to_minutes)

df.head()
```
